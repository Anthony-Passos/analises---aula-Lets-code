```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(
    r"3.1-data-sheet-udemy-courses-business-courses.csv", sep=","
)
df2 =  pd.read_csv(
    r"3.1-data-sheet-udemy-courses-design-courses.csv", sep=","
)
df3 =  pd.read_csv(
    r"3.1-data-sheet-udemy-courses-music-courses.csv", sep=","
)
df4 =  pd.read_csv(
    r"3.1-data-sheet-udemy-courses-web-development.csv", sep=","
)
```

Quantos cursos estão disponíveis na plataforma?

```python
dados = [df1, df2, df3, df4]
df = pd.concat(dados)
df
```

```python
cursos = df['course_title'].count()

print(f'Existem ao todo {cursos} Disponiveis')
```

Quais e quantos são os cursos que abordam o assunto: JavaScript?

```python
df['course_title'] = df['course_title'].str.lower()
df[df['course_title'].str.contains('javascript', na=False)]
```

Qual é o preço médio (coluna price) dos cursos oferecidos na plataforma?

```python
media_preco = np.round(df[['price']].mean(), 2)
media_business = np.round(df1[['price']].mean(), 2)
media_design = np.round(df2[['price']].mean(), 2)
media_music = np.round(df3[['price']].mean(), 2)
media_web = np.round(df4[['price']].mean(), 2)

x = ['Business', 'Design', 'Music', 'Web']
y = [media_business,media_design,media_music,media_web]

plt.plot(x, y, linestyle='solid')
plt.show()
print(f'O preço medio dos cursos em R$ oferecidos é {y}')
```

Quais são os cinco cursos com maior número de inscritos (coluna num_subscribers)?

```python
top5_mais_inscritos = df.sort_values(by='num_subscribers', ascending=False)
top5_mais_inscritos.head(5)
```

Dos cinco cursos mais populares em termos de inscritos, mostre o rate médio, máximo e mínimo de cada um deles.

```python
top5_mais_inscritos[['Rating']].apply(['min', 'max', 'mean'])
```

Apresente os dez cursos mais visualizados na plataforma.

```python
top10_cursos_avaliados = df.sort_values(by='num_reviews', ascending=False)
top10_cursos_avaliados.head(10)
```

A partir dos dez cursos mais vistos, mostre: 1) qual tem o maior número de inscritos; 2) qual tem o maior rate (avaliação do curso).

```python
inscricoes = top10_cursos_avaliados.sort_values(by='num_subscribers', ascending=False)
print(f"O curso com mais inscritos é {inscricoes['course_title'][0]} com {[inscricoes['num_subscribers'].max()]} inscritos")
```
